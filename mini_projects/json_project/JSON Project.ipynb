### Import packages into namespace

````python
# Import pandas and NumPy to access data structures and functions
import pandas as pd
import numpy as np

# Import json to encode/decode JSON data file(s)
import json

# Import json_normalize to flatten JSON file as a pandas DataFrame
from pandas.io.json import json_normalize

# Import matplotlib.pyplot for visualization
import matplotlib.pyplot as plt
````

## 1. Find the 10 countries with most projects.

First, load JSON file as a string using `json.load()`. Next, review contents of `json_str` to identify deeply nested data and determine best wasy to proceed with engineering data (not shown).

````python
# load json file as string
json_str = json.load((open('data/world_bank_projects.json')))
````

***
Upon reviewing nested JSON data file, flattened data as a pandas DataFrame `df_result` using `json_normalize()` along with parameters such as `record_path=` to parse nested arrays within nested array. Set index to `'countryshortname'` using `set_index()`.

````python
# normalize JSON file as DataFrame and parse 'mjtheme_namecode' 
# and include 'id' and 'countryshortname' columns 
df_result = json_normalize(json_str, meta=['countryshortname'], record_path='mjtheme_namecode')

# index DataFrame using 'countryshortname'
df_result.set_index(['countryshortname'], inplace=True)
````

***
Group DataFrame by `'countryshortname'` and count the number of projects for each country using `'count()'`.
Also, sort rows by `'code'` in descending order and select the Top 10 countries with most projects.
Dropped the column `'name'` since it is not needed.

**Note:** the group by, sort, and drop could have all been done in single command using the following:
```python
df_result.groupby(level=0).count().sort_values('code', ascending=False)[:10].drop(columns=['name'])
```

````python
# group by 'countryshortname' and count the number of projects for each country
df_countries_top10 = df_result.groupby(level=0).count()

# sort values in descending order and select the Top 10 countries with most projects
df_countries_top10 = df_countries_top10.sort_values('code', ascending=False)[:10]

# drop the 'name' column as it is not needed
df_countries_top10.drop(columns=['name'], inplace=True)

# alternate approach would be to combine above code into single line as follows:
#df_countries_top10 = df_result.groupby(level=0).count().sort_values('code', ascending=False)[:10].drop(columns=['name'])
````

***
Format columns and index labels and display **Top 10 Countries with Most Projects**.

````python
# assign label for columns and index 
df_countries_top10.columns.names = ['Number of Projects']
df_countries_top10.index.names = ['Country']

# rename the 'code' column to 'Total'
df_countries_top10.rename(columns={'code':'Total'}, inplace=True)

# display the DataFrame containing the 
# Top 10 Countries with Most Projects
df_countries_top10   # or can use print(df_countries_top10)
````

***
Visualize the **Top 10 Countries with Most Projects** as bar plot.
Here we can see that **Indonesia** has the most projects while **India** has the second most projects.

````python
# create bar plot using plot() with kind='bar'
df_countries_top10.plot(kind='bar', title='Top 10 Countries with Most Projects')

# label the X and Y ticks 
plt.ylabel('Number of Projects')
plt.xlabel('Countries')

# rotate the x tick labels 45 degrees and display bar plot
plt.xticks(rotation=45)
plt.show()
````

## 2. Find the top 10 major project themes 
(using column `'mjtheme_namecode'`)

Flatten the json_str data as a pandas DataFrame `df_result` using `json_normalize()` along with parameters such as `record_path=` to parse nested arrays within nested array.

While it wasn't necessary for **Exercise 2**, I replaced empty values for column `'name'` with `NaN`'s so I could apply forward fill to populate the null values (explained in **Ex. 3**).

Create a dictionary of Index : Values to rename the `'code'` column with actual project theme names.

````python
# normalize JSON file as DataFrame and parse 'mjtheme_namecode' 
df_result = json_normalize(json_str, record_path='mjtheme_namecode')

# select empty values for 'name' and store in boolean Series
null_data = df_result['name'] == ''

# replace empty values with NaN using numpy.nan
df_result.loc[null_data, 'name'] = np.nan

# forward fill NaN values with previous value
df_result.fillna(method='ffill', inplace=True)

# create a Key:Values dictionary containing Index:Values
# to be used for renaming the 'code' with project name
dict_result = dict(zip(df_result.code, df_result.name))
````

***
Group DataFrame by `'code'` and count the number of instances for each project theme using `'count()'`. Also, sort rows by `'name'` in descending order and select the **Top 10 Project Themes**.

````python
# group by 'code' and count the number of instance for each project theme
# sort values in descending order and select the Top 10 Project Themes
df_projects_top10 = df_result.groupby('code').count().sort_values('name', ascending=False)[:10]
````

***
Format columns and index labels as well as rename the project theme codes with project theme names and display **Top 10 Project Themes**.

````python
# assign label for columns and index 
df_projects_top10.columns.names = ['Major Themes']
df_projects_top10.index.names = ['Theme Name']

# rename the 'name' column to 'Total'
df_projects_top10.rename(columns={'name':'Total'}, inplace=True)

# rename the project codes with project theme names
df_projects_top10.rename(index=dict_result, inplace=True)

# display the DataFrame containing the 
# Top 10 Project Themes
df_projects_top10
````

***
Visualize the **Top 10 Project Themes** as bar plot.
Here we can see that the **Environment and natural resources management** them is the top project theme across all countries in the data set.

````python
# create bar plot using plot() with kind='bar'
df_projects_top10.plot(kind='bar', title='Top 10 Projects Themes')

# label the X and Y ticks 
plt.ylabel('Number of Instances')
plt.xlabel('Major Themes')

# rotate the x tick labels 45 degrees and display bar plot
#plt.xticks(rotation=45)
plt.show()
````

## 3. In Exercise 2, some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

Flatten the json_str data as a pandas DataFrame `df_result` using `json_normalize()` along with parameters such as `record_path=` to parse nested arrays within nested array.

Display information about the DataFrame to show how many null objects are present in the data set.

````python
# normalize JSON file as DataFrame and parse 'mjtheme_namecode' 
df_result = json_normalize(json_str, record_path='mjtheme_namecode')

# using info(), we can see that even though there are empty values in 'name' column,
# there are no null values which can impact analysis
df_result.info()
````

***
Replace empty values for column `'name'` with `NaN`'s using NumPy.nan so that I could apply forward fill to populate the null values. Sort DataFrame by values in `'code'` column.

After replacing empty/missing values in `'name'` column with NaN's, 
Display information about the DataFrame to show that there were empty values that have now been replaced with NaN's.

````python
# select empty values for 'name' and store in boolean Series
null_data = df_result['name'] == ''

# replace empty values with NaN using numpy.nan
df_result.loc[null_data, 'name'] = np.nan

# sort DataFrame by 'code' values
df_result.sort_values('code', inplace=True)

# after replacing empty values in 'name' column, we can use info() 
# to see that there are now 122 null values in 'name' column
df_result.info()
````

***
Apply Forward Fill (using `.fillna()`) to replace `NaN`'s with previous values in sorted `'name'` column.

After replacing `NaN`'s in `'name'` column with previous values, 
Display information about the DataFrame to show that there are no empty or NaN values.

````python
# forward fill NaN values with previous value
df_result.fillna(method='ffill', inplace=True)

# after applying forward fill to replace NaN with
# previous values, we can use info() to see that there  
# are zero empty values or null objects in 'name' column
df_result.info()
````
