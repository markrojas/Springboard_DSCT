# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

## B. Provide support for your arguments and your own recommendations with a statistically sound analysis:

* **Setup an appropriate hypothesis test.** 

A proposed hypothesis test that is appropiate for this study is that there is no difference in means for excess readmission rates between small capacatiy (<300 discharges) and not-so-small capacity (>300 discharges) hospitals/facilities. The alternate hypothesis, is that there is a significant difference in the means. 

$H_{0}: \mu_{\bar{A}} - \mu_{\bar{B}} = 0$

$H_{a}: \mu_{\bar{A}} - \mu_{\bar{B}} \neq 0$

Level of significance: ${α}$ = 0.01

We will use a **Two-sample T-test** to test significance.

* **Compute and report the observed significance value (or p-value).**

In this case, we want to test the hypothesis that small capacity (A) and not-so-small capacity (B) hospitals/facilities have the same mean excess readmission rates, but not necessarily the same distributions.

I will apply a bootstrap approach as a permutation test is not appropriate when distributions differ.

```python
def bootstrap_replicate_1d(data, func):
    return func(np.random.choice(data, size=len(data)))

def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data, func)

    return bs_replicates
```

```python
# set alpha
alpha = 0.01

# confirm preliminary report statistics
df = hospital_read_df[hospital_read_df['Number of Discharges'] != 'Not Available'].copy()
clean_df = df[df['Excess Readmission Ratio'].notnull()].copy()
clean_df.loc[:, 'Number of Discharges'] = clean_df['Number of Discharges'].astype(int)

a = clean_df[clean_df['Number of Discharges']<300]
b = clean_df[clean_df['Number of Discharges']>=300]

a_mean = round(np.mean(a['Excess Readmission Ratio']),3)
b_mean = round(np.mean(b['Excess Readmission Ratio']),3)
print('Hospitals/Facilities with number of discharges < 300 have a mean excess readmission rate of: {}'.format(a_mean))
print('Hospitals/Facilities with number of discharges > 300 have a mean excess readmission rate of: {}'.format(b_mean))

# compute difference of means
empirical_diff_means = a_mean - b_mean
print('Difference of Means: {}'.format(empirical_diff_means))

# compute mean of all excess readmission rates
mean_readmin = round(np.mean(clean_df['Excess Readmission Ratio']),3)
print('Mean excess readmission rate for all Hospitals/Facilities: {}'.format(mean_readmin))

# generate shifted arrays
a_shifted = a['Excess Readmission Ratio'] - np.mean(a['Excess Readmission Ratio']) + mean_readmin
b_shifted = b['Excess Readmission Ratio'] - np.mean(b['Excess Readmission Ratio']) + mean_readmin

# compute 10,000 bootstrap replicates from shifted arrays
bs_replicates_a = draw_bs_reps(a_shifted, np.mean, 10000)
bs_replicates_b = draw_bs_reps(b_shifted, np.mean, 10000)

# get replicates of difference of means
bs_replicates = bs_replicates_a - bs_replicates_b

ci = np.percentile(bs_replicates, [1, 99])
print('99% Confidence Interval: {}'.format(ci))

# compute the p-value
p_val = np.sum(bs_replicates>=empirical_diff_means) / len(bs_replicates)
print("\n******* Result ********\n")
if(p_val > alpha): 
    print('The p-value ({}) is > {}, so we cannot reject the null hypothesis.'.format(p_val, alpha))
    print('Thus, we are able to say that the means for excess readmission rates are equal for both small and not-so-small capacities.')
else:
    print('The p-value ({}) is <= {}, so we can reject the null hypothesis.'.format(p_val, alpha))
    print('Thus, we are unable to say that the means for excess readmission rates are equal for both small and not-so-small capacities.')
print("\n***********************\n")
```

* **Report statistical significance for  𝛼  = .01.**

The p-value is zero and thus, it is statistically significant.

* **Discuss statistical significance and practical significance. Do they differ here? How does this change your recommendation to the client?**

Based on the size of a p-value we get to either reject the null hypothesis (low p-value) or fail to reject the null hypothesis (high p-value). This is considered to be statistically significant. While practical significance sometimes goes hand in hand with statistical significance, this is not always the case. 

Practical significance arises out of applicability of the result in decision making. Practical significance is more subjective and depends upon external factors like cost, time, objectives, etc. apart from statistical significance.
